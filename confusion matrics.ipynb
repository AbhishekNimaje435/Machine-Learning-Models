# Confusion Matrix , Recall , Precision And F1_score

```python
# import required library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import placement dataset
data=pd.read_csv(r"C:\Users\HP\Downloads\placement-dataset.csv")
print(data.head(3))
```

```python
x=data.iloc[:,:-1]
y=data["placement"]
```

```python
# import train_test_split model
from sklearn.model_selection import train_test_split
```

```python
# split data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
# import LogisticRegression model
from sklearn.linear_model import LogisticRegression
```

```python
# call to the LogisticRegression()
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
# import confusion_matrix,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

```python
# call to the  confusion_matrix,
cf=confusion_matrix(y_test,lr.predict(x_test))
cf
```

```python
sns.heatmap(cf,annot=True)
plt.title("Confusion Metrics")
#plt.show()
```

```python
# accuracy of precision
precision_score(y_test,lr.predict(x_test))*100
```

```python
# accuracy of recall
recall_score(y_test,lr.predict(x_test))*100
```

```python
# accuracy of f1_score
f1_score(y_test,lr.predict(x_test))*100
```
