# Apriori Association Algorithm

```python
import pandas as pd
import numpy as np
pd.set_option("display.max_rows",500)
```

```python
data=pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Abhishek\groceries - groceries.csv")
data.head(1)
```

```python
market=[]
for i in range(0,data.shape[0]):
    cus=[]
    for j in data.columns:
        if type(data[j][i])==str:
            cus.append(data[j][i])
            market.append(cus)
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p=collections.Counter(l)
```

```python
d={"Item Name":p.keys(),"values":p.values()}
```

```python
#pd.DataFrame(d).sort_values(by=["values"],ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr=TransactionEncoder()
tr.fit(market)
df=pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df,min_support=0.09,use_colnames=True,max_len=3).sort_values(by=["support"])
```
