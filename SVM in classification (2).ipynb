```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\HP\Downloads\Placement (1).csv")
print(data.head(3))
```

```python
sns.scatterplot(x="CGPA",y="IQ",hue="Placement",data=data)
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel='linear')
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```
